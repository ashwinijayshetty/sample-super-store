```python
# Let's create a class named `display_all`, by which later we call on the next command.

def display_all(superstore_df):
    with pd.option_context("display.max_rows", 20, "display.max_columns", 20): 
        display(superstore_df)
```

```python
display_all(superstore_df.head(10).T)
```

```python
display_all(superstore_df.describe(include='all').T)
```
